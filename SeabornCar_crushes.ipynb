**Data Exploration**

```python
highest_total_row = car.loc[car['total'].idxmax()]
print(highest_total_row)
```

```python
lowest_total_row = car.loc[car['total'].idxmin()]
print(lowest_total_row)
```

```python
#Highest State
lowest_total = car['total'].min()
state_with_lowest_total = car.loc[car['total'].idxmin(), 'abbrev']
print(f"{state_with_lowest_total} has the lowest total car crashes with {lowest_total} crashes")

#Lowest State
highest_total = car['total'].max()
state_with_highest_total = car.loc[car['total'].idxmax(), 'abbrev']
print(f"{state_with_highest_total} has the highest total car crashes with {highest_total} crashes")
```

```python
#Another way to get the Highest and the Lowest State

#Highest State
#1. Sort the total in descending order
sort_descending_total = car.sort_values(by='total', ascending=False).head(1)

#2. Get the first row
highest_total_row = sort_descending_total.iloc[0]
highest_total_row[["abbrev", "total"]]
print(f"{highest_total_row['abbrev']} has the highest total car crashes with {highest_total_row['total']} crashes")
```

```python
#Lowest State
#1. Sort the total in ascending order
sort_ascending_total = car.sort_values(by='total', ascending=True).head(1)

#2. Get the first row
lowest_total_row = sort_ascending_total.iloc[0]
print(f"{lowest_total_row['abbrev']} has the lowest total car crashes with {lowest_total_row['total']} crashes")
```

```python
#Correlation
speed_corr = car['speeding'].corr(car['total'])
print("Correlation between 'speeding' and 'total':", speed_corr)

alcohol_corr = car['alcohol'].corr(car['total'])
print("Correlation between 'alcohol' and 'total':", alcohol_corr)
```

```python
#Top 10 states with insurance premium
# Sort by insurance premiums in descending order
top_10_premiums = car.sort_values(by="ins_premium", ascending=False).head(10)

# Display the result
print(top_10_premiums[["abbrev", "ins_premium"]])
```
